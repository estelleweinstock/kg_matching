## Plot Functions

```python
def plot_loss(data):
    
    plt.figure(figsize=(14, 6), dpi=80)

    ax = sns.lineplot(data=data["losses"], palette=['b'])
    ax.set(xlabel='Epochs', ylabel='Training Loss')
    ax.yaxis.get_label().set_fontsize(20)
    ax.xaxis.get_label().set_fontsize(20)
    ax.tick_params(axis="x", labelsize=15)
    ax.tick_params(axis="y", labelsize=15)
    ax.set(ylim=(0, 1))


    plt.show()

    
def plot_train_prec_rec(data):
    plt.figure(figsize=(14, 6), dpi=80)

    ax = sns.lineplot(data=data[["precisions_train","recalls_train"]], palette=['r', 'g'], linewidth=0.5, dashes=False)
    plt.legend(labels=["Precision","Recall"], fontsize=18)
    ax.set(xlabel='Epochs', ylabel='Precision / Recall')
    ax.yaxis.get_label().set_fontsize(20)
    ax.xaxis.get_label().set_fontsize(20)
    ax.tick_params(axis="x", labelsize=15)
    ax.tick_params(axis="y", labelsize=15)
    ax.set(ylim=(0, 1))

    plt.show()
```

```python
def plot_test_prec_rec(data):
    plt.figure(figsize=(14, 6), dpi=80)

    ax = sns.lineplot(data=data[["precisions_test","recalls_test"]], palette=['r', 'g'], linewidth=0.5, dashes=False)
    plt.legend(labels=["Precision","Recall"], fontsize=18)
    ax.set(xlabel='Epochs', ylabel='Precision / Recall')
    ax.yaxis.get_label().set_fontsize(20)
    ax.xaxis.get_label().set_fontsize(20)
    ax.tick_params(axis="x", labelsize=15)
    ax.tick_params(axis="y", labelsize=15)
    ax.set(ylim=(0, 1))

    plt.show()
```
