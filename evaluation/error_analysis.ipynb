## Plot Functions

```python
# central node
def plot(errors):
    plt.figure(figsize=(14, 6), dpi=80)

    ax = sns.lineplot(data=errors['proportion'])
    ax.set(xlabel='Proportion of matches',ylabel='Epochs with matches')
    ax.set(xlabel='Epochs', ylabel='Precision / Recall')
    ax.yaxis.get_label().set_fontsize(20)
    ax.xaxis.get_label().set_fontsize(20)
    ax.tick_params(axis="x", labelsize=15)
    ax.tick_params(axis="y", labelsize=15)
    ax.set(ylim=(0, 1))


    plt.show()
```

```python
# model guesses
def plot_amount(errors):
    plt.figure(figsize=(14, 6), dpi=80)

    ax = sns.lineplot(data=errors['lengths'])
    ax.set(xlabel='Epochs', ylabel='Amount of model guesses')
    ax.yaxis.get_label().set_fontsize(20)
    ax.xaxis.get_label().set_fontsize(20)
    ax.tick_params(axis="x", labelsize=15)
    ax.tick_params(axis="y", labelsize=15)
    ax.set(ylim=(0, 3500))


    plt.show()
```

## Data Preparation 

```python
lengths = []
most_common_el = []
for l in right:
    lengths.append(len(l))
    counter = Counter(l)
    try:
        most_common_el.append(counter.most_common(1))
    except:
        pass
```

```python
lens = lengths
```

```python
val = []
count = []
row = []
for el in most_common_el:
    try:
        val.append(el[0][0]) 
        count.append(el[0][1])
        row.append('')
    except: 
        row.append('')
```
